```python

titanic = sns.load_dataset("titanic")

df = titanic[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'alone']]


df['sex'] = df['sex'].map({'male': 0, 'female': 1})
df['alone'] = df['alone'].astype(int)

corr = df.corr()
corr['survived'].sort_values(ascending=False)
```

```python
corr['survived'].sort_values(ascending=False)
```

```python
corr = df.corr(method='pearson')
corr['survived'].sort_values(ascending=False)
```

```python
corr=df.corr()
corr['survived']
plt.figure(figsize=(11,8))
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr,mask=mask,vmin=-1, vmax=1, annot=True,
            cmap='RdYlBu_r', linewidths=2)
```

```python
# GDP per capita
economy = df2020[['Country name', 'GDP per capita']]
economy = (
    economy
    .groupby('Country name')['GDP per capita']
    .mean()
    .sort_values(ascending=False)
)

economy_rank = economy.rank(method='min', ascending=False)['South Korea']


# Social support
social = df2020[['Country name', 'Social support']]
social = social.groupby('Country name')['Social support'].mean().sort_values(ascending=False)
social_rank = social.rank(method='min', ascending=False)['South Korea']

# Life expectancy
health = df2020[['Country name', 'Healthy life expectancy']]
health = health.groupby('Country name')['Healthy life expectancy'].mean().sort_values(ascending=False)
health_rank = health.rank(method='min', ascending=False)['South Korea']

# Freedom
free = df2020[['Country name', 'Freedom to make life choices']]
free = free.groupby('Country name')['Freedom to make life choices'].mean().sort_values(ascending=False)
free_rank = free.rank(method='min', ascending=False)['South Korea']

```

```python
import numpy as np
import pandas as pd

happy_만족도 = pd.read_excel('data4/삶의만족도.xlsx')
happy_건강   = pd.read_excel('data4/건강.xlsx')
happy_경제   = pd.read_excel('data4/경제.xlsx')
happy_사회참여 = pd.read_excel('data4/사회참여.xlsx')
happy_교육   = pd.read_excel('data4/교육.xlsx')
happy_안전   = pd.read_excel('data4/안전.xlsx')
happy_여가   = pd.read_excel('data4/여가.xlsx')
happy_환경   = pd.read_excel('data4/환경.xlsx')
```

```python
happy_df = happy_만족도.copy()

```

```python
happy_df['건강'] = happy_건강['평균']
happy_df['경제'] = happy_경제['평균']
happy_df['사회참여'] = happy_사회참여['평균']
happy_df['교육'] = happy_교육['평균']
happy_df['안전'] = happy_안전['평균']
happy_df['여가'] = happy_여가['평균']
happy_df['환경'] = happy_환경['평균']

happy_df.rename(columns={'평균': '삶의 만족도'}, inplace=True)

```

```python
happy_df.head()
```

```python
happy_df['삶의 만족도'].fillna(
    happy_df['삶의 만족도'].mean(),
    inplace=True
)
```

```python
happy_df1 = happy_df.dropna()
happy_df.info()
happy_df1.info()

```

```python
happy_df1.describe()
```

```python

happy_df1.groupby('시도')[['삶의 만족도', '건강']].agg(
    mean_삶의만족도=('삶의 만족도', 'mean'),
    sum_건강=('건강', 'sum')
)



```

```python
happy_df1.groupby('시도').agg(
    삶의만족도_mean=('삶의 만족도', 'mean'),
    건강_mean=('건강', 'mean'),
    건강_median=('건강', 'median'),
    건강_sum=('건강', 'sum'),
    건강_std=('건강', 'std')
)
```
