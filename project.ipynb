```python
import pandas as pd
import folium
import json

# μ—‘μ…€ νμΌ λ΅λ“
df = pd.read_excel("./static/files/ν–‰λ³µμ§€μ(2021).xlsx")

print("π“ μ—‘μ…€ μ»¬λΌ λ©λ΅")
print(list(df.columns))

# π”Ή GeoJSONκ³Ό λ§¤μΉ­λ  μ»¬λΌ μ΄λ¦„ ν†µμΌ
df = df.rename(columns={"μ§€μ—­": "sido"})

print("\nπ“ λ³€κ²½λ μ»¬λΌ ν™•μΈ")
print(list(df.columns))

# GeoJSON λ΅λ“
geo_path = "./static/files/ctprvn.json"

with open(geo_path, encoding="utf-8") as f:
    geo = json.load(f)

print("\nπ“ GeoJSON ν¬ν•¨ μ§€μ—­λ…")
names = [x["properties"]["CTP_KOR_NM"] for x in geo["features"]]
print(names)

# μ§€λ„ μƒμ„±
m = folium.Map(location=[36.5, 127.8], zoom_start=7)

folium.Choropleth(
    geo_data=geo,
    data=df,
    columns=["sido", "ν–‰λ³µμ§€μ"],
    key_on="properties.CTP_KOR_NM",
    fill_color="PuBuGn",
    fill_opacity=0.9,
    line_opacity=0.3,
    legend_name="μ§€μ—­λ³„ ν–‰λ³µμ§€μ"
).add_to(m)

m
```
