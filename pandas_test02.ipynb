```python
# 중복 제거
iris_df = iris_df.drop_duplicates()

# 종별 통계
stats_df = iris_df.groupby('species').agg(
    sepal_length_mean=('sepal_length', 'mean'),
    sepal_width_mean=('sepal_width', 'mean'),
    petal_length_mean=('petal_length', 'mean'),
    petal_width_mean=('petal_width', 'mean')
)

stats_df
```

```python
iris_df.groupby('species').mean()
```

```python
import matplotlib.pyplot as plt
iris_df.sepal_length[:10].plot(kind='bar')
plt.title("Sepal Length Visualation")
plt.xlabel("Data")
plt.ylabel("Sepal Length")
plt.show()
```

```python
df2=iris_df.groupby('species').mean()
print(df2)
df2.sepal_length[:].plot(kind='bar',rot=0)
plt.title('Sepal Length Visualation')
plt.xlabel('data')
plt.ylabel('selal length mean')
plt.show()
```
