```python
from folium import plugins
N = 100 
data = np.array( 
[
np.random.uniform(low=35, high=38, size=N),
np.random.uniform(low=125, high=129, size=N),
] ).T 
popups = [str(i) for i in range(N)] 
m = folium.Map([37.566345, 126.977893], zoom_start=8) 
plugins.MarkerCluster(data, popups=popups).add_to(m) 
m.save(os.path.join('data', 'Plugins_1.html')) 
m
```

```python
df=pd.read_csv('data8/CCTV_20190917.csv')
```

```python
df1=df.loc[df['카메라대수']>=1,:]
```

```python
# df1['소재지도로명주소']=df1['소재지도로명주소'].apply(str.replace(' ',''))
popups=df1['소재지도로명주소'].tolist()
popups
```

```python
lat_avg=np.array(df1['위도'].tolist()).mean()
lon_avg=np.array(df1['경도'].tolist()).mean()
m=folium.Map(location=[lat_avg, lon_avg])
```

```python
import numpy as np
import folium

data = np.array([df1['위도'].tolist(), df1['경도'].tolist()]).T

m = folium.Map(
    location=[data[0][0], data[0][1]],
    zoom_start=12,
    width=800,
    height=800
)

for i in range(len(data)):
    folium.Marker(
        [data[i][0], data[i][1]],
        popup=popups[i],
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(m)

m
```
